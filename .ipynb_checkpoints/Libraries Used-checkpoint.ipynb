```python
# Matplotlib
# Data Visualisation: presentation of data in a pictorial or graphical format
# Matplotlib is a visualisation library for 2D plots of array
# It gives lowlevel API that gives precise control of how things look and where they are displayed
import matplotlib.pyplot as plt
xpoints= (17,22,26,35)
ypoints= (12,15,34,40) 
```

```python
# Plotting line graph
plt.plot(xpoints, ypoints)
plt.show() 
```

```python
# Marking points
plt.plot(xpoints, ypoints, 'o')
plt.show() 
```

```python
# End is the marker it can be circle, star, plus shape, etc. 
# The size can be changed as well and change its colour also 
# 143 defined colours in matplotlib
plt.plot(xpoints, ypoints, '-o')
plt.show()
```

```python
plt.plot(xpoints, ypoints, marker='*', ms='15')
plt.show()
```

```python
plt.plot(xpoints, ypoints, marker='*', ms='15', mec='r', mfc='y')
plt.show() 
```

```python
# change line style
plt.plot(xpoints, ypoints, ls='dashed')
plt.show() 
```

```python
plt.plot(xpoints, ypoints, ls='dashed', color='#445566')
plt.show() 
```

```python
plt.plot(xpoints, ypoints, ls='dashed', color='#445566', lw='4')
plt.show() 
```

```python
# change figure size
plt.figure(figsize=(7,7))
plt.plot(xpoints, ypoints, ls='dashed', color='#445566', lw='4')
plt.show() 
```

```python
# add labels and titles
plt.plot(xpoints, ypoints, ls='dashed', color='#445566', lw='4')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Data')
plt.show() 
```

```python
# can apply grids
plt.grid()
plt.plot(xpoints, ypoints, ls='dashed', color='#445566', lw='4')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Data')
plt.show() 
```

```python
plt.grid(axis='x')
plt.plot(xpoints, ypoints, ls='dashed', color='#445566', lw='4')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Data')
plt.show() 
```

```python
# Scatter
x = (25,27,28,31,32,45,38,67,45)
y = (20,39,66,21,30,55,14,69,30)
plt.scatter(x,y)
plt.show() 
```

```python
x = (25,27,28,31,32,45,38,67,45)
y = (20,39,66,21,30,55,14,69,30)
plt.scatter(x,y)
x = (15,17,18,16,12)
y = (56,22,44,18,11)
plt.scatter(x,y)    # automatically shows in 2 different colors
plt.show() 
```

```python
x = (25,27,28,31,32,45,38,67,45)
y = (20,39,66,21,30,55,14,69,30)
plt.scatter(x,y)
x = (15,17,18,16,12)
y = (56,22,44,18,11)
plt.scatter(x,y, color='hotpink') # can change color
plt.show() 
```

```python
# Bars
x= ('P',"Q",'R',"S")
y = (13,28,11,20)
plt.bar(x,y)
plt.show() 
```

```python
# reduce width
x= ('P',"Q",'R',"S")
y = (13,28,11,20)
plt.bar(x,y, width=0.2)
plt.show() 
```

```python
# horizontal bar
x= ('P',"Q",'R',"S")
y = (13,28,11,20)
plt.barh(x,y)
plt.show() 
```

```python
# reduce height
x= ('P',"Q",'R',"S")
y = (13,28,11,20)
plt.barh(x,y, height=0.2)
plt.show() 
```

```python
# Pie charts
y = (23,89,22,56,67)
plt.pie(y)
plt.show()
```

```python
# add labels
y = (23,89,22,56,67)
chart = ('q','r','s','t','u')
plt.pie(y, labels=chart)
plt.show() 
```

```python
# add values
y = (23,89,22,56,67)
chart = ('q','r','s','t','u')
val = (0.3,0,0,0.5,0)
plt.pie(y, labels=chart, explode=val)
plt.show()
```

```python
# Seaborn
# Predominantly used for making statistical graphics
# Depends on numpy, scipy, matplotlib, pandas, statsmodels
# Datasets implicitly supported by seaborn libraries:-
# https://github.com/mwaskom/seaborn-data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Dataset loading
a = sns.load_dataset("flights")
# relplot combines scatter and line plots together
sns.relplot(x='month', y='passengers', data=a) 
```

```python
a = sns.load_dataset("flights")
sns.relplot(x='month', y='passengers', data=a, hue='year') 
```

```python
a = sns.load_dataset("flights")
sns.relplot(x='month', y='passengers', data=a, hue='year', kind='line') 
```

```python
# Plotting with categorical data
# Dividing data in discrete groups
sns.catplot(x='month', y='passengers', data=a) 
```

```python
# can use box, boxen, swarm, strip in kind
sns.catplot(x='month', y='passengers', data=a, kind='violin')
```

```python
# univariate and bivariate distributions
# sns.distplot(x="month",y="passengers",data=a,hue="year")  # ERROR
```

```python
# Multiplot grids
# Graphs plotted side by side with same axes' scale for aid
b = sns.PairGrid(a)
b.map(plt.scatter)
sns.pairplot(a)
```

```python
# Plot aesthetics
# Used to make interface more user friendly and customizing themes with high level interfaces
sns.set(style='dark')       # white, whitegrid, darkgrid
sns.relplot(x='month', y='passengers', data=a)
```
