```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Reading the dataset in mv variable
mv = pd.read_csv('netflix_titles.csv')
```

```python
# First 5 data only
mv.head()   
```

```python
# All Info about the table [structure]
mv.info()
```

```python
# Generating heat map [white line show null values]
sns.heatmap(mv.isnull())
```

```python
# Data cleaning [removing null values]
# inplace=True means data is updated and nothing is returned
mv['country'].replace(np.nan, 'No data', inplace=True)
mv['cast'].replace(np.nan, 'No data', inplace=True)
mv['director'].replace(np.nan, 'No data', inplace=True)

# dropna() is used to remove missing values
mv.dropna(inplace=True)
# drop_duplicates() is used to remove duplicate values from dataset 
mv.drop_duplicates(inplace=True)
# isnull().sum() returns the number of missing values in the data set
mv.isnull().sum()

# Showing data
mv.info()
mv
```

```python
# Extracting Date information and adding them as new columns in the end
mv['date_added'] = pd.to_datetime(mv['date_added'])
mv['month_name_added'] = mv['date_added'].dt.month_name()
mv['month_added'] = mv['date_added'].dt.month
mv['year_added'] = mv['date_added'].dt.year
mv
```

```python
# Data Analysis
# Separating TV shows and Movies
shows = mv[mv['type']=='TV Show']
movies = mv[mv['type']=='Movie']
movies.shape, shows.shape
```

```python
# Pie chart
category = ['TV Shows', 'Movies']
# taking .shape[0] for total rows
count = [shows.shape[0], movies.shape[0]]
fig = plt.figure(figsize=(10,7))
# for % values use autopct
plt.pie(count, labels=category, autopct="%0.2f")
plt.show()
```

```python
# There are multiple countries also see 7 below
mv['country'].head(20)
```

```python
# Show primary country in this case [1st country]
mv['country_f'] = mv['country'].apply(lambda x: x.split(',')[0])
mv['country_f'].head(10)
```

```python
# Shortening country names to avoid overlap
mv['country_f'].replace('United States', 'USA', inplace=True)
mv['country_f'].replace('United Kingdom', 'UK', inplace=True)
mv['country_f'].replace('South Korea', 'S. Korea', inplace=True)
mv['country_f']
```

```python
# Count total wrt to countries
mv['country_f'].value_counts()
```

```python
# value_counts() give unique values
# reset_index() is done to reset multilevel index 0 to n-1
mv['country_f'].value_counts().reset_index()
```

```python
# Renaming above
mv['country_f'].value_counts().reset_index().rename(columns={"index":"Country", "country_f":"count"})
```

```python
# Drop No data
country_count = mv['country_f'].value_counts().reset_index().rename(columns={"index":"Country", "country_f":"count"})
country_count.drop(2, inplace=True)
country_count
```

```python
# Top 10 countries [slicing]
top_countries = country_count[:10]
top_countries.head(10)
```

```python
# Bar chart 
fig = plt.figure(figsize=(17,7))
sns.barplot(x=top_countries['Country'], y=top_countries['count'])
```

```python
# Bar chart and pie chart together
fig = plt.figure(figsize=(17,7))
# Subplot 1
plt.subplot(1,2, 1)
sns.barplot(x=top_countries['Country'], y=top_countries['count'])
# Subplot 2
plt.subplot(1,2, 2)
plt.pie(top_countries['count'], labels=top_countries['Country'], autopct="%0.2f")
plt.show()
```

```python
# Amount of content added as function of time
col = 'year_added'
total_count = mv[col].value_counts().reset_index().rename(columns={'index':'year', 'year_added':'count'})
movie_count = movies[col].value_counts().reset_index().rename(columns={'index':'year', 'year_added':'count'})
show_count = shows[col].value_counts().reset_index().rename(columns={'index':'year', 'year_added':'count'})
total_count
```

```python
fig = plt.figure(figsize=(15,5))
plt.xticks(np.arange(2008, 2022, 1 ))
sns.lineplot(data = total_count, x = 'year', y = 'count')
sns.lineplot(data = movie_count, x = 'year', y = 'count')
sns.lineplot(data = show_count, x = 'year', y = 'count')
plt.legend(['Total', 'Movies', 'Shows'])
plt.show()
```

```python
# arranges 1 to 10 [excluded] in interval of 3
np.arange(1, 10, 3) 
```

```python
# Content ratings
plt.figure(figsize=(12,9))
sns.countplot(x='rating', data=mv, hue='type')
```

```python
# month in which most content was released
month_order=['January', 'February', 'March', 'April', 'May','June','July','August','September','October','November','December']
sns.countplot(x='month_name_added', data=mv, order=month_order)
sns.set_style('whitegrid')
plt.xticks(rotation = 90)
```

```python
# Actor with most content
mv['cast']
cast_df = mv['cast'].str.split(',', expand=True)
cast_df
```

```python
actor = cast_df.melt(value_name = 'Actor')['Actor']
actor = actor.value_counts().reset_index().rename(columns = {'index':'Actor', 'Actor':'count'})
actor.drop(0, inplace=True)
actor
```

```python
top = actor[:10]
sns.catplot(x = 'count', y='Actor', data=top)
plt.xticks(np.arange(22, 40, 1), rotation=90)
plt.show()
```

```python
# Indian Movie directors with most movies
movies_ind = mv[mv['country_f']=='India']
movies_ind = movies_ind[movies_ind['type']=='Movie']
ind_directors = movies_ind['director'].value_counts().reset_index().rename(columns={'index':'director', 'director':'count'})
ind_directors.drop(0, inplace=True)
dir = ind_directors[:15] 
plt.barh(dir['director'], dir['count'])
plt.xticks(np.arange(0,10,1))
```
